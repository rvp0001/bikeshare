```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def train_print_summary(column_list,X_train,y_train):
    X_train_lm = X_train[column_list]
    X_train_lm = sm.add_constant(X_train_lm)    
    lr = sm.OLS(y_train, X_train_lm).fit()
    print(lr.summary())
    vif = pd.DataFrame()
    vif['Features'] = X_train_lm.columns
    vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(X_train_lm.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
```

```python
train_print_summary(['temp', 'tm'],X_train,y_train)
```

```python
train_print_summary(['hum','temp','windspeed','weathersit_1','weathersit_2','season_1','season_2','season_3','workingday','tm'],X_train,y_train)
```
